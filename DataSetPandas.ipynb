# Now we will use MATPLOTLIB AS PAYPLOT for data visulaization 

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data=pd.read_csv("C:/PSL_Season_09_dataset.csv")
```

```python
data.columns
```

```python
df = pd.DataFrame(data)
df['total_runs'] = df['runs_off_bat'] + df['extras']
total_runs_by_team = df.groupby('batting_team')['total_runs'].sum().reset_index()
plt.figure(figsize=(10, 6))
plt.bar(total_runs_by_team['batting_team'], total_runs_by_team['total_runs'], color='skyblue')
plt.title('Total Runs Scored by Each Batting Team')
plt.xlabel('Batting Team')
plt.ylabel('Total Runs')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
df = pd.DataFrame(data)
plt.figure(figsize=(10, 6))
plt.scatter(df['runs_off_bat'], df['extras'], color='blue', alpha=0.6)
plt.title('Scatter Plot of Runs Off Bat vs Extras')
plt.xlabel('Runs Off Bat')
plt.ylabel('Extras')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
df = pd.DataFrame(data)
extras_by_team = df.groupby('bowling_team')[['wides', 'noballs', 'byes', 'legbyes', 'penalty']].sum().reset_index()
extras_by_team.set_index('bowling_team').plot(kind='bar', stacked=True, figsize=(10, 6), colormap='tab20c')
plt.title('Distribution of Extras by Bowling Team')
plt.xlabel('Bowling Team')
plt.ylabel('Total Extras')
plt.legend(title='Type of Extras')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
df = pd.DataFrame(data)
df['total_runs'] = df['runs_off_bat'] + df['extras']
total_runs_by_season = df.groupby('season')['total_runs'].sum().reset_index()
plt.figure(figsize=(10, 6))
plt.plot(total_runs_by_season['season'], total_runs_by_season['total_runs'], marker='o', linestyle='-', color='b')
plt.title('Total Runs Scored Over Different Seasons')
plt.xlabel('Season')
plt.ylabel('Total Runs')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
df = pd.DataFrame(data)
runs_by_player = df.groupby('striker')['runs_off_bat'].sum().reset_index()
top_players = runs_by_player.sort_values(by='runs_off_bat', ascending=False).head(10)
plt.figure(figsize=(12, 6))
plt.bar(top_players['striker'], top_players['runs_off_bat'], color='green')
plt.title('Top Players with Most Runs')
plt.xlabel('Player')
plt.ylabel('Total Runs')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
df = pd.DataFrame(data)
wickets_df = df[df['wicket_type'].notnull() | df['other_wicket_type'].notnull()]
wickets_by_bowler = wickets_df['bowler'].value_counts().reset_index()
wickets_by_bowler.columns = ['bowler', 'wickets']
top_bowlers = wickets_by_bowler.sort_values(by='wickets', ascending=False).head(10)
plt.figure(figsize=(12, 6))
plt.bar(top_bowlers['bowler'], top_bowlers['wickets'], color='blue')
plt.title('Top Bowlers with Most Wickets')
plt.xlabel('Bowler')
plt.ylabel('Total Wickets')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
df = pd.DataFrame(data)
player_stats = df.groupby('striker').agg({
    'runs_off_bat': 'sum',
    'ball': 'count' 
}).reset_index()
player_stats['strike_rate'] = (player_stats['runs_off_bat'] / player_stats['ball']) * 100
top_10_scorers = player_stats.sort_values(by='runs_off_bat', ascending=False).head(10)
plt.figure(figsize=(14, 7))
plt.bar(top_10_scorers['striker'], top_10_scorers['strike_rate'], color='orange')
plt.title('Strike Rate of Top 10 Run Scorers')
plt.xlabel('Player')
plt.ylabel('Strike Rate')
plt.xticks(rotation=45)
plt.ylim(0, max(top_10_scorers['strike_rate']) + 10)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

# LINE OF REGRESSION 

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np
data=pd.read_csv("C:/PSL_Season_09_dataset.csv")
df = pd.DataFrame(data)
X = df[['ball']] 
y = df['runs_off_bat']  
model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)
plt.figure(figsize=(10, 6))
plt.scatter(df['ball'], df['runs_off_bat'], color='blue', label='Actual Runs')
plt.plot(df['ball'], y_pred, color='red', linewidth=2, label='Linear Regression Line')
plt.title('Linear Regression: Balls Faced vs Runs Scored')
plt.xlabel('Balls Faced')
plt.ylabel('Runs Scored')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
print(f"Intercept: {model.intercept_}")
print(f"Coefficient: {model.coef_[0]}")
```
